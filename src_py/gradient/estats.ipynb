```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import epanet as epa
import epamodule as em
import seaborn as sns
from cjsbot import CjsBot
from scipy import optimize
from scipy import stats
import time
import os 
```

```python
seeds = ['308', '343', '491', '661', '769']
for seed in seeds:
    sd = []
    mode_value = []
    mode_count = []
    for i in range(1,16):
        arq = open(f'./teste_nos/{seed}/{i}/dists.txt')
        dists = np.loadtxt(arq)
        arq.close()
        sd.append(dists.std())
        mode_value.append(stats.mode(dists).mode[0])
        mode_count.append(stats.mode(dists).count[0]/dists.shape[0])
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), sd)
    plt.ylabel('Desvio Padrão')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_nos/{seed}/std.png', format='png')
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), mode_value)
    plt.ylabel('Moda')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_nos/{seed}/moda_valores.png', format='png')
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), mode_count)
    plt.ylabel('% de valores que são moda')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_nos/{seed}/moda_countv2.png', format='png')
```

```python
seeds = ['308', '343', '491', '661', '769']
for seed in seeds:
    sd = []
    mode_value = []
    mode_count = []
    count = []
    mode_count_abs = []
    for i in range(1,16):
        arq = open(f'./teste_conjuntos/{seed}/{i}/dists.txt')
        dists = np.loadtxt(arq)
        arq.close()
        sd.append(dists.std())
        mode_value.append(stats.mode(dists).mode[0])
        mode_count.append(stats.mode(dists).count[0]/dists.shape[0])
        mode_count_abs.append(stats.mode(dists).count[0])
        count.append(dists.shape[0])
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), sd)
    plt.ylabel('Desvio Padrão')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_conjuntos/{seed}/std.png', format='png')
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), mode_value)
    plt.ylabel('Moda')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_conjuntos/{seed}/moda_valores.png', format='png')
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), mode_count)
    plt.ylabel('% de valores que são moda')
    plt.xlabel('Dimensão')
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_conjuntos/{seed}/moda_countv2.png', format='png')
    
    fig, ax = plt.subplots()
    ax.plot(np.arange(1,16), mode_count_abs)
    ax.plot(np.arange(1,16), count)
    plt.ylabel('nº de pontos')
    plt.xlabel('Dimensão')
    # colocar legenda
    #plt.title('Desvio padrão da distância\ em função da dimensão')
    fig.savefig(f'./teste_conjuntos/{seed}/certo.png', format='png')
```
