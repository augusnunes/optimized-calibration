```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
target = np.array([0.075, 0.812, 0.317, 0.581, 0.752, 0.994, 0.967, 0.511, 0.851,
              0.925, 0.842, 0.295, 0.633, 0.522, 0.306])
```

```python
def dist(x, target):
    return scipy.linalg.norm(x-target)
```

```python
std = []
dist0 = []
c = []
media = []
for i in range(1,16):
    distancias = np.loadtxt(f'./{i}/dists.txt')
    unique, counts = np.unique(distancias, return_counts=True)
    estat = dict(zip(unique, counts))
    if 0.0 in estat:
        dist0.append(estat[0.0]/len(distancias)*100)
    else:
        dist0.append(0)
    std.append(np.array(distancias).std())
    media.append(np.array(distancias).mean())

```

```python
media
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1,16),dist0)
plt.xlabel("Dimensão")
plt.ylabel("Quantidade (%)")
fig.savefig('t1_dim_dist0.png', format='png')
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1,16),std)
plt.xlabel("Dimensão")
plt.ylabel("Desvio Padrão")
fig.savefig('t1_dim_std.png', format='png')
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(1,16),media)
media = np.array(media)
std = np.array(std)
ax.fill_between(np.arange(1,16), media - std, media + std, alpha=0.2)
plt.xlabel("Dimensão")
plt.ylabel("Média da Distância")
fig.savefig('t1_dim_mean.png', format='png')
```

```python
i = 1
pontos = np.loadtxt(f'./{i}/y.txt')
distancias = []
for j in range(pontos.shape[0]):
    if i == 1:
        distancias.append(dist(pontos[j], target[:i]))
    else:
        distancias.append(dist(pontos[j,:], target[:i]))
```
