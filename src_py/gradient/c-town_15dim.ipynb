# 4 Dimensões [0.001, 0.2]

```python
links = ["../../networks/c-town/nodes", 
         "../../networks/c-town/links", 
         "../../networks/c-town/rede.inp", 
         "../../networks/c-town/4dim.csv"]
```

```python
t = np.array([0.01,0.079,0.115,0.199])
```

```python
t.shape
```

```python
rv = epa.RealValues(links, t)
rv.target_nodes
rv.getRealValue()
```

```python
net = epa.Rede(links, rv.groups, t, [0.001,0.2])
```

```python
net.objetivo(net.target)
```

```python
pontos = np.random.random((100,4))*0.2 +0.001
```

```python
x = []
y = []
for i in range(100):
    try:
        o = optimize.minimize(net.objetivo, pontos[i,:], method='Nelder-Mead')
        x.append(pontos[i,:])
        y.append(np.round(o.x, decimals=3))
    except:
        continue
CjsBot().send_message("Acabou")
```

```python
d = []
for i in range(len(y)):
    d.append(net.get_dist(x[i]))
    
```
