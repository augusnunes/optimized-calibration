```python
moda_diferent0 = []
for seed in seeds:
    for posicao_no in posicao_nos:
        for q_no in q_nos:
            for dim in range(1,16):
                if (df[(df.seed == seed) & (df.posicao_no == posicao_no) & (df.quantidade_nos==q_no) & (df.dimensao == dim)].moda != 0).values[0]:                  
                    v = df[(df.seed == seed) & (df.posicao_no == posicao_no) & (df.quantidade_nos==q_no) & (df.dimensao == dim)].moda.values[0]
                    moda_diferent0.append([seed, posicao_no, q_no, dim, v])
                    break
```

```python
moda_diferent0 = np.array(moda_diferent0)
```
