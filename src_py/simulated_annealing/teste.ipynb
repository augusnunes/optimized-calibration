```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
arq = open('teste/dist.txt')
d = np.loadtxt(arq)
arq.close()
```

```python
arq = open('teste/f.txt')
f = np.loadtxt(arq)
arq.close()
```

```python
df = pd.DataFrame({'dist':d, 'f':f})
```

```python
df1 = df[df.dist<0.5]
```

```python
plt.scatter(df1.dist, df1.f)
plt.xlabel("Distância")
plt.ylabel('f(x_bsf)')
plt.yscale('log')
plt.title('Simulated Annealing\nDimensões=14 e 40 nós avaliados')
```
